```python
X = flights[['Airline', 'Source', 'Destination', 'Dep_Time', 'Total_Stops', 'Date_of_Journey']]
y = flights['Price']
```

```python
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
```

```python
X
```

```python
X['day'] = pd.to_datetime(X['Date_of_Journey'], format = '%d/%m/%Y').dt.day
X['month'] = pd.to_datetime(X['Date_of_Journey'], format = '%d/%m/%Y').dt.month
X['year'] = pd.to_datetime(X['Date_of_Journey'], format = '%d/%m/%Y').dt.year
```

```python
X.drop('Date_of_Journey', inplace=True, axis=1)
```

```python
X.head()
```

```python
X['Total_Stops'].value_counts()
```

```python
def stops(n):
    if n == '1 stop':
        return 1
    elif n == '2 stops':
        return 2
    elif n == '3 stops':
        return 3
    elif n == '4 stops':
        return 4
    else:
        return 0
X['Stops'] = X['Total_Stops'].apply(stops)
```

```python
X.drop('Total_Stops', inplace=True, axis=1)
```

```python
X.head()
```

```python
X['Dep_Hour'] = pd.to_datetime(X['Dep_Time']).dt.hour
X['Dep_Mins'] = pd.to_datetime(X['Dep_Time']).dt.minute
```

```python
X.drop(['Dep_Time', 'year'], axis=1, inplace=True)
```

```python
X.head()
```

```python
X['Airline'].value_counts()
```

```python
air = pd.get_dummies(X[['Airline']], drop_first = True) 
```

```python
len(X['Airline'].value_counts())
```

```python
src = pd.get_dummies(X[['Source']], drop_first = True) 
dest = pd.get_dummies(X[['Destination']], drop_first = True) 
```

```python
x = X.iloc[:, [3,4,5,6,7]]
Xi = pd.concat([air, src, dest, x], axis=1, join='inner')
```

```python
Xi.head()
```

```python
X_train, X_test, y_train, y_test = train_test_split(Xi, y, test_size=0.1)
print(X_train.shape, y_train.shape, X_test.shape, y_test.shape)
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import cross_val_score
```

```python
from sklearn.ensemble import RandomForestRegressor
tree2 = RandomForestRegressor(n_estimators=1000, criterion='squared_error', max_depth=None, min_samples_split=24, min_samples_leaf=1, min_weight_fraction_leaf=0.0, max_features='auto', max_leaf_nodes=None, min_impurity_decrease=0.0, bootstrap=True, oob_score=False, n_jobs=None, random_state=None, verbose=0, warm_start=False, ccp_alpha=0.0, max_samples=None)
```

```python
tree2.fit(X_train, y_train)
tree2.score(X_test, y_test)*100
```

```python
from sklearn.metrics import accuracy_score, precision_score, mean_squared_error
y_pred = tree2.predict(X_test) 
```

```python
lsp = pd.DataFrame({'Actual' : y_test, 'Predicted' : y_pred}) 
```
